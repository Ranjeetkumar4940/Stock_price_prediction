```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# 1. Data Preparation
def prepare_data(df):
    df['Date'] = pd.to_datetime(df['Date'])
    
    df.set_index('Date', inplace=True)
    
    df.fillna(method='ffill', inplace=True)
    df.fillna(method='bfill', inplace=True)
    
   
    for col in ['Open', 'High', 'Low', 'Close', 'Volume']:
        q_low = df[col].quantile(0.01)
        q_high = df[col].quantile(0.99)
        df = df[(df[col] >= q_low) & (df[col] <= q_high)]
    
    return df
```

```python
# 2. Exploratory Data Analysis (EDA)
def eda_pipeline(df, stock_name="Stock"):
    print(f"Performing EDA for {stock_name}")
    
    
    print("\nSummary Statistics:")
    print(df.describe())
    
    print("\nMissing Values:")
    print(df.isnull().sum())
    
    # Plot Price Trends (Open, High, Low, Close)
    plt.figure(figsize=(12, 6))
    df[['Open', 'High', 'Low', 'Close']].plot(title=f"{stock_name} - Price Trends")
    plt.ylabel("Price")
    plt.xlabel("Date")
    plt.show()

    # Plot Trading Volume
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=df, x=df.index, y="Volume", label="Volume")
    plt.title(f"{stock_name} - Trading Volume")
    plt.ylabel("Volume")
    plt.xlabel("Date")
    plt.legend()
    plt.show()

    # Rolling Mean (50 and 200 days)
    df[['Close', 'Rolling_Mean_50', 'Rolling_Mean_200']].plot(figsize=(12, 6), title=f"{stock_name} - Price and Moving Averages")
    plt.ylabel("Price")
    plt.xlabel("Date")
    plt.show()

    # Seasonality Decomposition 
    decomposition = seasonal_decompose(df['Close'], model='additive', period=251)  # Assuming 252 trading days in a year
    decomposition.plot()
    plt.show()

    # Correlation Heatmap
    plt.figure(figsize=(8, 6))
    sns.heatmap(df[['Open', 'High', 'Low', 'Close', 'Volume', 'Rolling_Mean_50', 'Rolling_Mean_200', 'Pct_Change', 'Log_Return', 'RSI']].corr(), annot=True, cmap='coolwarm')
    plt.title(f"{stock_name} - Correlation Heatmap")
    plt.show()
    
    return df
```

```python
# 3. Feature Engineering
def feature_engineering(df):
    # Lagged Features (Previous Day's Price)
    df['Prev_Close'] = df['Close'].shift(1)  # Previous day's close price
    df['Prev_Open'] = df['Open'].shift(1)    # Previous day's open price

    # Rolling Mean (50 and 200 days)
    df['Rolling_Mean_50'] = df['Close'].rolling(window=50).mean()
    df['Rolling_Mean_200'] = df['Close'].rolling(window=200).mean()

    # Percentage Change from the previous day
    df['Pct_Change'] = df['Close'].pct_change()

    # Log Return 
    df['Log_Return'] = np.log(df['Close'] / df['Close'].shift(1))

    # Bollinger Bands
    df['BB_Middle'] = df['Close'].rolling(window=20).mean()
    df['BB_Upper'] = df['BB_Middle'] + 2 * df['Close'].rolling(window=20).std()
    df['BB_Lower'] = df['BB_Middle'] - 2 * df['Close'].rolling(window=20).std()

    # Relative Strength Index (RSI)
    delta = df['Close'].diff(1)
    gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
    rs = gain / loss
    df['RSI'] = 100 - (100 / (1 + rs))

    # Drop missing values generated due to rolling calculations
    df.dropna(inplace=True)
    
    return df
```

## Interpreting Bollinger Bands:

Squeeze: When bands are very close together, it indicates low volatility. It may be a precursor to a sharp price move (breakout or breakdown).
Breakouts: Prices breaking through the bands often signal strong momentum, but they do not indicate the direction of the move.
Reversal Potential: When prices are at the extremes of the bands, they often revert toward the middle band (mean reversion).

```python
for stock_name, df in dfs.items():
    print(f"Processing {stock_name}...")
    
    # Data Preparation
    df = prepare_data(df)  # Clean data and handle anomalies
    
    # Feature Engineering
    df = feature_engineering(df)  # Create new features
    
    # EDA (Exploratory Data Analysis)
    dfs[stock_name] = eda_pipeline(df, stock_name=stock_name)  # Perform EDA
```

# ADF TEST

  * Null hypothesis- Time series is non-stationary
  * If p value of test < 0.05 then we can reject the null hypothesis
  * if p value > 0.05 need to find the order of differencing

```python
from statsmodels.tsa.stattools import adfuller

result=adfuller(df.dropna())
print(f"ADF Statistics: {result[0]}")
print(f"p-value: {result[1]}")
```

# Conclusion from test :

  * p-value is very high than 0.05 we need to do order of differencing

```python
import matplotlib.pyplot as plt
import numpy as np
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
```

```python
# Define the plot function
def plot_actual_vs_predicted(actual, predicted):
    plt.figure(figsize=(10,6))
    plt.plot(actual, label='Actual Values', color='blue')
    plt.plot(predicted, label='Predicted Values', color='red', linestyle='dashed')
    plt.title('Actual vs Predicted Values')
    plt.xlabel('Time')
    plt.ylabel('Stock Price')
    plt.legend()
    plt.show()
```

```python
# The function to predict using XGBoost
def xgb_predict(train, val):
    train = np.array(train)
    X, y = train[:, :-1], train[:, -1]
    model = XGBRegressor(objective="reg:squarederror", n_estimators=100)
    model.fit(X, y)

    val = np.array(val).reshape(1, -1)
    pred = model.predict(val)
    return pred[0]
```
